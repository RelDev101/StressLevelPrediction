## Exploratory Data Analysis

```python
import pandas as pd

df = pd.read_csv("../data/SaYoPillow.csv")

df
```

```python
df.rename(columns={'sr': 'snoring rate', 
                   'rr':'respiration rate', 
                   't': 'body temperature', 
                   'lm':'limb movement',
                   'bo':'blood oxygen', 
                   'rem':'eye movement rate', 
                   'sr.1':'sleeping hours', 
                   'hr':'heart rate',
                   'sl':'stress levels'}, inplace=True)

df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['stress levels'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='stress levels', data=df)
plt.title('Distribusi Stress Levels')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df, hue='stress levels')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='stress levels', y='heart rate', data=df)
plt.title("Heart Rate vs Stress Levels")
plt.show()
```

```python
df.drop('stress levels', axis=1).hist(figsize=(12, 10), bins=20)
plt.suptitle("Distribusi Fitur-Fitur")
plt.show()
```
